```python
import  numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from glob import glob
```

# Take The First Image Of Male

```python
malepath = glob('./data/Male Faces/*.jpg')
```

```python
len(malepath) # length of male path
```

```python
path = malepath[0] # Take the first image to perform the work
```

```python
img = cv2.imread(path) # Read the image
```

```python
plt.imshow(img) # Show the image
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert the BGR image to gray
```

```python
plt.imshow(gray, cmap = 'gray') # Show the BGR image
```

```python
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml') # use xml files for detect the img
```

```python
faces = haar.detectMultiScale(gray, scaleFactor=1.5, minNeighbors=5) # Create the coordinates of img
print(faces)
```

```python
for x,y,w,h in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0),5) # detect the face of image
plt.imshow(img)    
    
```

```python
crop_img = img[y:y+h, x:x+h] # crop the image
```

```python
plt.imshow(crop_img) # Show the crop image
```

```python
cv2.imwrite('f_1 (1).jpg', crop_img) # save the crop img into folder
```

```python
def extract_images(path, gender, i): # Extract And Save all Image in Folder
    img = cv2.imread(path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = haar.detectMultiScale(gray, 1.5,4)
    for x,y,w,h in faces:
        roi = img[y:y+h, x:x+w]
        if gender == 'female':
            cv2.imwrite('./data/crop/female_crop/{}_{}.png'.format(gender,i),roi)
        else:
              cv2.imwrite('./data/crop/male_crop/{}_{}.png'.format(gender,i),roi)
```

```python
extract_images(malepath[0], 'male',1) # Extract and save one image in folder
```

```python
for i, path in enumerate(malepath): # Extract and save all image in Folder
    try:
        extract_images(path, 'male Faces', i)
        print('INFO:{}/{} processed succesfully'.format(i, len(malepath)))
    except:
         print('INFO:{}/{} cannot be processed'.format(i,len(malepath)))
```
