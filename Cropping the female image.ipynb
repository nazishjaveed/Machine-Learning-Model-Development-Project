```python
img = cv2.imread(path) # Read The Image
```

```python
plt.imshow(img) # Show The Read Image
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert the BGR image to gray
```

```python
plt.imshow(gray, cmap = 'gray') # Show the BGR image
```

```python
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml') # use xml files for detect the img
```

```python
faces = haar.detectMultiScale(gray, scaleFactor=1.5, minNeighbors=4) # Create the coordinates of img
print(faces)
```

```python
for x,y,w,h in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0),5) # detect the face of image
plt.imshow(img)    
    
    
```

```python
def extract_images(path, gender, i): # Extract And Save all Image in Folder
    img = cv2.imread(path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = haar.detectMultiScale(gray, 1.5,4)
    for x,y,w,h in faces:
        roi = img[y:y+h, x:x+w]
        if gender == 'male':
            cv2.imwrite('./data/crop/male_crop/{}_{}.png'.format(gender,i),roi)
        else:
              cv2.imwrite('./data/crop/female_crop/{}_{}.png'.format(gender,i),roi)
            
        
```

```python
extract_images(femalepath[0], 'female',1) # Extract and save one image in folder
```

```python
for i, path in enumerate(femalepath): # Extract and save all image in Folder
    try:
        extract_images(path, 'Female Faces', i)
        print('INFO:{}/{} processed succesfully'.format(i, len(femalepath)))
    except:
         print('INFO:{}/{} cannot be processed'.format(i,len(femalepath)))
        
      
            
```
