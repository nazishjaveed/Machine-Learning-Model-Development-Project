# Import the Library

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

# Read the Image

```python
img = cv2.imread('img4.jpg')
```

```python
img.shape
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
print(gray.shape)
```

```python
plt.imshow(gray, cmap = 'gray')
```

# Resize the Image

```python
img_resize = cv2.resize(gray, (500,400), cv2.INTER_AREA)
```

```python
plt.imshow(img_resize, cmap = 'gray')
```

# Enlarge the Image

```python
img_en = cv2.resize(gray , (2000,1600), cv2.INTER_CUBIC)
```

```python
img_en.shape
```

```python
plt.imshow(img_en, cmap = 'gray')
```
